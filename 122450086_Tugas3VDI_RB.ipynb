3. Membuat Visualisasi

Berikut adalah contoh kodenya untuk membuat visualisasi data penjualan bulanan menggunakan matplotlib dan seaborn.

Data Simulasi: Data penjualan bulanan disimulasikan untuk empat produk (Product A, Product B, Product C, Product D) selama 12 bulan.

Grafik Line Plot: Digunakan untuk memperlihatkan tren penjualan setiap produk sepanjang tahun.

Warna Berbeda: Setiap produk diwakili oleh warna yang berbeda untuk memudahkan perbandingan antar produk.

Marker: Ditambahkan marker (titik) pada garis untuk memperjelas perubahan penjualan di tiap bulan.

Dengan visualisasi ini, pemilik toko dapat langsung melihat bulan-bulan ketika penjualan meningkat, produk mana yang paling diminati, dan bagaimana pola penjualan berubah sepanjang tahun.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Simulasi data penjualan bulanan untuk beberapa produk
months = pd.date_range(start='2023-01-01', periods=12, freq='M')
products = ['Product A', 'Product B', 'Product C', 'Product D']

# Membuat dataframe simulasi
data_sales = pd.DataFrame({
    'Month': np.tile(months, len(products)),
    'Product': np.repeat(products, len(months)),
    'Sales': np.random.randint(500, 3000, size=len(products) * len(months)),
})

# Membuat plot untuk visualisasi data penjualan
plt.figure(figsize=(10, 6))
sns.lineplot(data=data_sales, x='Month', y='Sales', hue='Product', marker='o')

# Menambahkan judul dan label sumbu
plt.title('Monthly Sales for Different Products in 2023')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()

# Menampilkan grafik
plt.show()
```

Bar plot cocok jika ingin membandingkan total penjualan antar produk di bulan-bulan tertentu atau total keseluruhan untuk tiap produk selama periode waktu.
Visualisasi ini cocok untuk membandingkan penjualan bulanan tiap produk secara berdampingan.

Bar plot digunakan untuk melihat penjualan pada setiap bulan dengan membandingkan panjang batang yang mewakili jumlah penjualan untuk masing-masing produk. Ini memudahkan pengguna untuk melihat total perbandingan penjualan di setiap bulan.

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data_sales, x='Month', y='Sales', hue='Product')

# Menambahkan judul dan label sumbu
plt.title('Monthly Sales Comparison for Different Products')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Jika ingin melihat pola penjualan secara keseluruhan, heatmap sangat berguna untuk melihat intensitas penjualan di setiap bulan dan produk.
visualisasi ini untuk melihat pola intensitas penjualan tiap produk selama 12 bulan.


Heatmap memvisualisasikan intensitas penjualan di setiap bulan untuk tiap produk dengan menggunakan warna. Semakin gelap warna, semakin tinggi penjualannya, sehingga Anda bisa dengan mudah mengidentifikasi bulan-bulan dengan penjualan tertinggi.

```python
# Mengubah data menjadi format pivot untuk heatmap
data_pivot = data_sales.pivot(index='Month', columns='Product', values='Sales')

# Membuat heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(data_pivot, annot=True, fmt=".0f", cmap="YlGnBu")

# Menambahkan judul
plt.title('Monthly Sales Heatmap for Each Product')
plt.xlabel('Product')
plt.ylabel('Month')
plt.tight_layout()
plt.show()
```

Pie chart cocok untuk menggambarkan distribusi penjualan setiap produk sebagai bagian dari total penjualan di periode tertentu (misalnya, sepanjang tahun atau di satu bulan tertentu).
Pada contohnya visualisasi ini enggambarkan persentase kontribusi penjualan setiap produk di bulan Januari.

Pie chart memberikan gambaran distribusi penjualan setiap produk sebagai bagian dari total penjualan pada bulan tertentu. Ini berguna untuk melihat seberapa besar kontribusi masing-masing produk.

```python
# Filter data untuk bulan Januari
january_sales = data_sales[data_sales['Month'] == '2023-01-31']

# Membuat pie chart
plt.figure(figsize=(6, 6))
plt.pie(january_sales['Sales'], labels=january_sales['Product'], autopct='%1.1f%%', startangle=90)

# Menambahkan judul
plt.title('Product Sales Distribution in January 2023')
plt.tight_layout()
plt.show()
```

Box plot sangat cocok untuk melihat distribusi penjualan tiap produk, terutama jika Anda ingin melihat variasi dan outlier.
Visualisasi ini nampilkan distribusi penjualan tiap produk sepanjang tahun.

Box plot menunjukkan rentang distribusi penjualan tiap produk sepanjang tahun. Ini sangat membantu untuk melihat variasi penjualan dan mengidentifikasi produk yang memiliki penjualan sangat tinggi (outliers).


```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_sales, x='Product', y='Sales')

# Menambahkan judul dan label sumbu
plt.title('Sales Distribution for Each Product in 2023')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

Setiap visualisasi memiliki kekuatan masing-masing tergantung pada tujuan analisis. Jika ingin mencari pola tren, line plot atau heatmap adalah pilihan terbaik. Jika ingin membandingkan penjualan total, bar plot atau stacked bar plot bisa lebih jelas. Untuk distribusi, pie chart dan box plot cocok digunakan.
Maka setiap visualisasi harus disesuaikan dengan kebutuhan, tujuan dan juga fungsi kepada user.
